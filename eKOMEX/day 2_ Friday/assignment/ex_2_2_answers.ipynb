## Create a simple line plot with two lines comparing the following measurements

## Style this plot

## Visualize the following election data with a paired bar chart

```python
parties = ["CDU","SPD","AfD", "FDP", "DIE LINKE", "GRÃœNE", "CSU", "Sonstige"]
result_2017 = [26.8,20.5,12.6,10.7,9.2,8.9,6.2,5.0]
result_2013 = [34.1,25.7,4.7,4.8,8.6,8.4,7.4,6.2]
```

```python
plt.figure(figsize=(10,5))
plt.bar(np.arange(len(parties)) -0.2, result_2013, width=0.4, label='2013')
plt.bar(np.arange(len(parties)), result_2017, width=0, tick_label=parties, )
plt.bar(np.arange(len(parties)) +0.2, result_2017, width=0.4, color="red", label='2017')
plt.legend()
```

## Create two pie charts for the above election data

```python
plt.figure(figsize=(5,5))
plt.pie(result_2013, labels=parties, colors = ["black", "red", "#4444FF", "yellow", "#BB0055", "green", "#1111CC","grey" ], counterclock=False, startangle=90)
plt.figure(figsize=(5,5))
plt.pie(result_2017, labels=parties, colors = ["black", "red", "#4444FF", "yellow", "#BB0055", "green", "#1111CC","grey" ], counterclock=False, startangle=90)
plt.show()
```
